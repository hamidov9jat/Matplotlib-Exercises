Instead of len(car_sales) try to use `car_sales.shape[0]` to get the number of rows.
It is more efficient than using `len` function

*Explanation*:
From a performance perspective, using the shape attribute of the dataframe is generally the most efficient way to get the number of rows in a pandas dataframe. This is because the shape attribute is an attribute of the dataframe object itself, so it does not require any additional computation to retrieve the value.

On the other hand, using the len function to get the number of rows in a dataframe requires iterating through the rows of the dataframe, which can be slower if the dataframe is large.

```python
# Add a date column
# car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=car_sales.shape[0])
car_sales
```

```python
# Make a bit more of a complicated plot

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"], cmap='bwr')

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(), title="Target")
# ax.set_facecolor('lightgrey')
ax.set_facecolor('lightgreen')
fig.set_facecolor('lightgrey')



# Add a meanline
ax.axhline(y=over_50["chol"].mean(),
           linestyle="--", color='#ff008c', linewidth=2);
```

### Adding another plot to existing styled one

```python
# Setup plot (2 rows, 1 column)
fig, (ax0, ax1) = plt.subplots(nrows=2, # 2 rows
                               ncols=1, 
                               sharex=True, 
                               figsize=(10, 8))

# Add data for ax0
scatter = ax0.scatter(over_50["age"], 
                      over_50["chol"], 
                      c=over_50["target"])
# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.set_facecolor('lightgreen')

# Setup a mean line
ax0.axhline(y=over_50["chol"].mean(), 
            color='b', 
            linestyle='--', 
            label="Average", linewidth=2)


# Add data for ax1
scatter = ax1.scatter(over_50["age"], 
                      over_50["thalach"], 
                      c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.set_facecolor('lightgreen')

# Setup a mean line
ax1.axhline(y=over_50["thalach"].mean(), 
            color='b', 
            linestyle='--', 
            label="Average", linewidth=2)

# Title the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```

*Note*:  The `legend_elements` method is only available for `PathCollection` objects, which are returned by the `scatter` function. It is not available for `Line2D` objects, which are returned by the `axhline` function.

To include the 'Average' label for the horizontal line in the legend of the plot, you can use the `legend` function of the `AxesSubplot` object and pass it the handle and label for the horizontal line directly.

```python
# Change cmap and horizontal line to be a different colour
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"], 
                     cmap="winter")
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

average_line = ax.axhline(y=over_50["chol"].mean(), 
           color='r', 
           linestyle='--', 
           label="Average");

# Get the handles and labels for the scatter plot and the horizontal line
scatter_handles, scatter_labels = scatter.legend_elements()

# ax.legend(*scatter.legend_elements(), title="Target")

# Use the legend function of the AxesSubplot object to display the legend
ax.legend(scatter_handles + [average_line], scatter_labels + [average_line.get_label()], title="Legend")
```
